#### Problem 4:
A palindromic number reads the same both ways. The largest palindrome made from the product of two 2-digit numbers is 9009 = 91 × 99.

Find the largest palindrome made from the product of two 3-digit numbers.

```python
def HighestPalindrome(low, high):
    
    highest_pal = 0 
    
    for x in range(high, low, -1):
        for y in range(high, low, -1):
            product = x * y
            if str(product) == str(product)[::-1]:
                if product > highest_pal:
                    highest_pal = product
    
    return highest_pal
```

What is the smallest positive number that is evenly divisible by all of the numbers from 1 to 20?

```python
import fractions 
  
# Returns the lcm of first n numbers 
def lcm(n): 
    ans = 1    
    for i in range(1, n + 1): 
        ans = (ans * i)/fractions.gcd(ans, i)   
    return ans 
  
n = 20
lcm(n)
```

#### Problem 6:

The sum of the squares of the first ten natural numbers is,

1^2 + 2^2 + ... + 10^2 = 385  

The square of the sum of the first ten natural numbers is,  

(1 + 2 + ... + 10)^2 = 55^2 = 3025

Hence the difference between the sum of the squares of the first ten natural numbers and the square of the sum is 3025 − 385 = 2640.

Find the difference between the sum of the squares of the first one hundred natural numbers and the square of the sum.

```python
def Sum_of_Squares(number):
    
    sum_of_squares = 0
    
    for i in range(1, number+1):
        sum_of_squares += i*i
    
    return sum_of_squares

def Square_of_Sums(number):
    
    sum_of_range = 0
    for i in range(1, number+1):
        sum_of_range += i
    
    return sum_of_range*sum_of_range

def Difference(number):
    return(Square_of_Sums(number) - Sum_of_Squares(number))

    
```

#### Problem 7: 
By listing the first six prime numbers: 2, 3, 5, 7, 11, and 13, we can see that the 6th prime is 13.

What is the 10 001st prime number?

```python
def isprime(n):
    '''check if integer n is a prime'''

    # 0 and 1 are not primes
    if n < 2:
        return 0

    # 2 is the only even prime number
    if n == 2: 
        return 1   

    # all other even numbers are not primes
    if not n & 1: 
        return 0

    # range starts with 3 and only needs to go up 
    # the square root of n for all odd numbers
    for x in range(3, int(n**0.5) + 1, 2):
        if n % x == 0:
            return 0
    return 1
```

```python
def find_x_prime(n):
    "Finds the X'th prime number"
    
    primes = []

    x = 1 

    while len(primes) != n:
        if isprime(x) == 1:
            primes.append(x)
            x += 1
        else:
            x+=1

    return(primes[-1])
```

```python
find_x_prime(10001)
```
