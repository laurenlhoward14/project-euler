### Functions that may help:
#### Fibonacci

````python
def fibonacci(n):
    if n <= 1:
        return n
    else:
        return(fibonacci(n-1) + fibonacci(n-2))
````

#### Prime Number Checker

````python
def isprime(n):
    '''check if integer n is a prime'''

    # 0 and 1 are not primes
    if n < 2:
        return 0

    # 2 is the only even prime number
    if n == 2: 
        return 1   

    # all other even numbers are not primes
    if not n & 1: 
        return 0

    # range starts with 3 and only needs to go up 
    # the square root of n for all odd numbers
    for x in range(3, int(n**0.5) + 1, 2):
        if n % x == 0:
            return 0
    return 1
````

#### Get Factors of a number

````python
def GetFactors(n):
    factor_values = []
    for i in range(1, n + 1):
        if n % i == 0:
            factor_values.append(i)
    return factor_values
````

#### Problem 3: Largest Prime Factor
The prime factors of 13195 are 5, 7, 13 and 29.

What is the largest prime factor of the number 600851475143 ?

````python
def MaxPrimeFactor(n):
    
    largest_prime_factor = 1
    
    # Remove any factors of 2: 
    while n % 2 == 0: 
        largest_factor = 2
        n = n/2
    
    # Look at odd factors
    p = 3 
    while n !=1:
        while n % p == 0:
            largest_factor = p
            n = n/p
            
        p+=2
    
    return largest_factor

    
````

#### Problem 4: Largest Palindromic Product
A palindromic number reads the same both ways. The largest palindrome made from the product of two 2-digit numbers is 9009 = 91 × 99.

Find the largest palindrome made from the product of two 3-digit numbers.

````python
def HighestPalindrome(low, high):
    
    highest_pal = 0 
    
    for x in range(high, low, -1):
        for y in range(high, low, -1):
            product = x * y
            if str(product) == str(product)[::-1]:
                if product > highest_pal:
                    highest_pal = product
    
    return highest_pal
````

#### Problem 5: Smallest Multiple
2520 is the smallest number that can be divided by each of the numbers from 1 to 10 without any remainder.

What is the smallest positive number that is evenly divisible by all of the numbers from 1 to 20?

````python
# Returns the lcm of first n numbers 
def lcm(n): 
    ans = 1    
    for i in range(1, n + 1): 
        ans = (ans * i)/fractions.gcd(ans, i)   
    return ans 
  
n = 20
lcm(n)
````

#### Problem 6: Sum Square Difference

The sum of the squares of the first ten natural numbers is,

1^2 + 2^2 + ... + 10^2 = 385  

The square of the sum of the first ten natural numbers is,  

(1 + 2 + ... + 10)^2 = 55^2 = 3025

Hence the difference between the sum of the squares of the first ten natural numbers and the square of the sum is 3025 − 385 = 2640.

Find the difference between the sum of the squares of the first one hundred natural numbers and the square of the sum.

````python
def Sum_of_Squares(number):
    
    sum_of_squares = 0
    
    for i in range(1, number+1):
        sum_of_squares += i*i
    
    return sum_of_squares

def Square_of_Sums(number):
    
    sum_of_range = 0
    for i in range(1, number+1):
        sum_of_range += i
    
    return sum_of_range*sum_of_range

def Difference(number):
    return(Square_of_Sums(number) - Sum_of_Squares(number))

    
````

````python
Difference(100)
````

#### Problem 7: 10001st Prime
By listing the first six prime numbers: 2, 3, 5, 7, 11, and 13, we can see that the 6th prime is 13.

What is the 10 001st prime number?

````python
def find_x_prime(n):
    "Finds the X'th prime number"
    
    primes = []

    x = 1 

    while len(primes) != n:
        if isprime(x) == 1:
            primes.append(x)
            x += 1
        else:
            x+=1

    return(primes[-1])
````

#### Problem 9: Special Pythagorean Triplet

A Pythagorean triplet is a set of three natural numbers, a < b < c, for which,

$a^2$ + $b^2$ = $c^2$  
For example: $3^2 + 4^2 = 5^2$

There exists exactly one Pythagorean triplet for which $a + b + c = 1000$.  
Find the product abc.

````python
# Multiply each number by each number - Arbitrary high number
for a in range(1,500):
    for b in range(1,500):
        
        # Only if b is greater than a as defined problem find c from c_sqrd
        if b > a: 
            c_sqrd = (a*a)+(b*b)
            c = np.sqrt(c_sqrd)
            
            # Sqrt of c has to be an integer
            if c.is_integer():
                sum_ints = a + b + c
                
                # If the sum of these integers equal 1000 then we found a b and c and can find the product
                if sum_ints == 1000:
                    print(f"a={a}, b={b}, c={c}")
                    print(f"Product: {a*b*c}")
                
                
````

````python
def HighestProds(sample):
    
    highest_prods = 0

    for i in range(len(sample)):
        for j in range(len(sample)):

            # Across
            try: 
                across = sample[i][j]*sample[i][j+1]*sample[i][j+2]*sample[i][j+3]
                if across > highest_prods:
                    highest_prods = across
            except IndexError: 
                None

            # Down
            try:
                down = sample[i][j]*sample[i+1][j]*sample[i+2][j]*sample[i+3][j]
                if down > highest_prods:
                    highest_prods = down
            except IndexError: 
                None

            # Left to Right Diagnols
            try:
                ltor_diag = sample[i][i]*sample[i+1][j+1]*sample[i+2][j+2]*sample[i+3][j+3]
                if ltor_diag > highest_prods:
                    highest_prods = ltor_diag
            except IndexError:
                None

            # Right to Left Diagnols
            try:
                rtol_diag = sample[i][j]*sample[i+1][j-1]*sample[i+2][j-2]*sample[i+3][j-3]
                if rtol_diag > highest_prods:
                    highest_prods = rtol_diag
            except IndexError:
                None


    return highest_prods
````

The sequence of triangle numbers is generated by adding the natural numbers. So the 7th triangle number would be 1 + 2 + 3 + 4 + 5 + 6 + 7 = 28. The first ten terms would be:

1, 3, 6, 10, 15, 21, 28, 36, 45, 55, ...

Let us list the factors of the first seven triangle numbers:

`1: 1  
3: 1,3  
6: 1,2,3,6  
10: 1,2,5,10  
15: 1,3,5,15  
21: 1,3,7,21  
28: 1,2,4,7,14,28`  
We can see that 28 is the first triangle number to have over five divisors.

What is the value of the first triangle number to have over five hundred divisors? 76576500

````python
def TriangleNumber(n):
    "Finds nth triangle number"
    
    triangle = 0
    for num in range(1, n+1):
        triangle += num
    
    return triangle
````

#### Problem 14: Longest Collatz sequence

The following iterative sequence is defined for the set of positive integers:

n → n/2 (n is even)  
n → 3n + 1 (n is odd)

Using the rule above and starting with 13, we generate the following sequence:

13 → 40 → 20 → 10 → 5 → 16 → 8 → 4 → 2 → 1  
It can be seen that this sequence (starting at 13 and finishing at 1) contains 10 terms. Although it has not been proved yet (Collatz Problem), it is thought that all starting numbers finish at 1.

Which starting number, under one million, produces the longest chain?

NOTE: Once the chain starts the terms are allowed to go above one million.



````python
def GenerateSequence(n):
    "Generates Sequence defined by above rules for any number n"
    sequence = []

    number_to_test = n

    while number_to_test != 1:
        sequence.append(number_to_test)
        if number_to_test % 2 == 0:
            # number is even
            number_to_test = int(number_to_test/2)

        else:
            # number is odd
            number_to_test = int((3*number_to_test) +1)

    sequence.append(int(1)) # As the while statement will stop the sequence at 2
    return(len(sequence))
````

````python
longest_sequence = 0
length_of_sequence = 0

# Perform function for every number up to 1 million
# only update longest_sequence and length if a greater one is found
for n in range(1,1000001):
    if GenerateSequence(n) > length_of_sequence:
        length_of_sequence = GenerateSequence(n)
        longest_sequence = n 

print(f"""Starting Number: {longest_sequence} 
Length of Sequence: {length_of_sequence}""")
````

#### Problem 15: Lattice Paths
Starting in the top left corner of a 2×2 grid, and only being able to move to the right and down, there are exactly 6 routes to the bottom right corner.

How many such routes are there through a 20×20 grid?

````python
# This works for small grids - very slow for larger grids
def PathGenerator(gridsize):
    path = []
    counter = 0
    for i in range(gridsize):
        path.append(0) # Moves right
        path.append(1) # Moves down
    print(path)
    
    # Find possible permutations of this path and only count unique solutions
    solutions = []
    for i in itertools.permutations(path, gridsize*2):
        if i not in solutions:
            solutions.append(i)
        
    return(len(solutions))
````

#### Problem 21: Amicable Numbers

Let d(n) be defined as the sum of proper divisors of n (numbers less than n which divide evenly into n).
If d(a) = b and d(b) = a, where a ≠ b, then a and b are an amicable pair and each of a and b are called amicable numbers.

For example, the proper divisors of 220 are 1, 2, 4, 5, 10, 11, 20, 22, 44, 55 and 110; therefore d(220) = 284. The proper divisors of 284 are 1, 2, 4, 71 and 142; so d(284) = 220.

Evaluate the sum of all the amicable numbers under 10000

````python
def sum_of_factors(x):
   # Takes a number and sums its factors
    factors = []
    
    for i in range(1, x + 1):
        if x % i == 0:
            factors.append(i)
    return sum(factors[:-1])
````

````python
total = 0

for x in range(10001):
    factors_sum = sum_of_factors(x)
    
    if sum_of_factors(factors_sum) == x:
        # Check it's not the same number
        if x != factors_sum:
            total += x
        else:
            None
    else:
        None

print(total)
````

#### Problem 22: Names scores

Using names.txt (right click and 'Save Link/Target As...'), a 46K text file containing over five-thousand first names, begin by sorting it into alphabetical order. Then working out the alphabetical value for each name, multiply this value by its alphabetical position in the list to obtain a name score.

For example, when the list is sorted into alphabetical order, COLIN, which is worth 3 + 15 + 12 + 9 + 14 = 53, is the 938th name in the list. So, COLIN would obtain a score of 938 × 53 = 49714.

What is the total of all the name scores in the file?

````python
def names_list(filepath):
    """Reads in text file and returns names in list format"""
    text = open(filepath, 'r')
    contents = text.read()

    names_list = contents.split(",")

    for idx, name in enumerate(names_list):
        names_list[idx] = name.strip('\"')

    return names_list
````

````python
names = names_list("p022_names.txt")
names.sort()
````

````python
alphabet = 'ABCDEFGHIJKLMNOPQRSTUVWXYZ'

full_list_score = 0

for idx, name in enumerate(names):
    name_score = 0
    for letter in name:
        name_score += alphabet.index(letter)+1 # Calculate each letter score
    full_list_score += name_score * (idx+1) # Multiplied by position in list

print(full_list_score)
````

#### Problem 23: Non-Abundant Sums:

A perfect number is a number for which the sum of its proper divisors is exactly equal to the number. For example, the sum of the proper divisors of 28 would be 1 + 2 + 4 + 7 + 14 = 28, which means that 28 is a perfect number.

A number n is called deficient if the sum of its proper divisors is less than n and it is called abundant if this sum exceeds n.

As 12 is the smallest abundant number, 1 + 2 + 3 + 4 + 6 = 16, the smallest number that can be written as the sum of two abundant numbers is 24. By mathematical analysis, it can be shown that all integers greater than 28123 can be written as the sum of two abundant numbers. However, this upper limit cannot be reduced any further by analysis even though it is known that the greatest number that cannot be expressed as the sum of two abundant numbers is less than this limit.

Find the sum of all the positive integers which cannot be written as the sum of two abundant numbers.

````python
def GetFactors(n):
    factor_values = []
    for i in range(1, n + 1):
        if n % i == 0:
            factor_values.append(i)
    return factor_values
````

````python
# Find all abundant numbers up until 28123
abundant_number = []

for i in range(28124):
    
    if sum(GetFactors(i)[:-1]) > i:
        abundant_number.append(i)
    else:
        None
````

````python
product_of_abundant_nums = set() # Calculate all numbers that can be a product of abundant numbers

for i, x in enumerate(abundant_number): #add each element to each element in list
    for j in abundant_number[i:]:
        sum_of_elements = x+j 
        product_of_abundant_nums.add(sum_of_elements)     #sets discard duplicates automatically
````

````python
# Check all numbers not in that list and sum them
sum_of_nums_not_prod_of_abundant_nums = 0

for i in range(28124):
    if i not in product_of_abundant_nums:
        sum_of_nums_not_prod_of_abundant_nums +=i

sum_of_nums_not_prod_of_abundant_nums
````
